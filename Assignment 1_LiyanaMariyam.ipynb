```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Reading dataset into python

```python
data = pd.read_excel('iris.xls')
```

#### Displaying Columns in the dataset

```python
data.columns
```

#### Calculating mean of each column

```python
data.mean()
```

#### Checking for null values present

```python
data.isna().sum()
```

#### Pie chart of Classification:

```python
Count = data['Classification'].value_counts()
label = data['Classification'].unique()
plt.pie(Count,labels=label, autopct='%1.1f%%')
plt.show()
```

#### Bar graph of Classification v/s PL:

```python
plt.bar(data['Classification'], data['PL'])
plt.legend()
plt.show()
```

#### Box plot of the data:

```python
data1 = data.set_index('Classification')
plt.boxplot(data1)
plt.show()
```

#### Pairplot between different values in data:

```python
sns.pairplot(data)
plt.show()
```
